### Berechnen einer neuen Zielspalte

```python
# Berechnung einer neuen Zielspalte
# Veränderung des Goldpreises in den nächsten x (= predictdays) Tagen

class MyDataProcessor(): #TransformerMixin, BaseEstimator):
    def __init__(self, trainingsdays=5, predictdays=5):
        self.trainingsdays = trainingsdays
        self.predictdays = predictdays

    def fit(self, Xname='Open',column_list=None):
        self.column_list_ = column_list
        #self.df_ = df
        self.Xname_ = Xname
        return self

    def transform(self, df):
        df_x = df[[self.Xname_]].rename(columns={self.Xname_:'X0'})
        for n in range(1, self.trainingsdays):
            df_x.loc[:,'dX'+str(-n)] = df_x.X0.shift(n-1) / df_x.X0.shift(n) -1
        if self.column_list_:
            df_x.loc[:, self.column_list_] = df.loc[:,self.column_list_]
        df_x.loc[:,'y'] = df_x.X0.shift(-self.predictdays) / df_x.X0 - 1
        # df_x.dropna(inplace=True)
        self.X_ = df_x.drop('y', axis=1)
        self.y_ = df_x.loc[:,['y']]
        return self
```

```python
processor = MyDataProcessor(trainingsdays=0, predictdays=5)
processor.fit()
processor = processor.transform(df)

df["y"] = processor.y_
df["y"].tail(10)
```

```python
# Durch die Berechnung des neuen Zielwerts y mit vorausgehenden Datensätzen 
# entstehen in den letzten Datensätzen einige NaN-Werte.
# Diese Datensätze müssen gelöscht werden.
```
